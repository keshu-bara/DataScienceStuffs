### LineChart

```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
# create a line chart, years on x-axis , gdp on y-axis
plt.plot(years, gdp, color = 'green', marker='o', linestyle='solid')
#         x    , y ,                                 

#adding title
plt.title("Nominal GDP")

#add a label to the ya-axis
plt.ylabel("Billions of $")

#add a label to the x-axis
plt.xlabel("Years")

# Annotate each point with its value
for x, y in zip(years, gdp):
    plt.text(x, y + 0.2 , f'{y}',ha = 'center',fontsize = 9, color = 'red')

# used to adjust subplots
plt.tight_layout()

plt.show()
```

### Bar Charts

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
# bars are by default width 0.8, so we'll add 0.1 to the left coordinates
# so that each bar is centered

# x-axis = movies and y-axis = num_oscars

plt.bar(movies,num_oscars,color = 'blue')
# plt.bar(movies,num_oscars)
plt.ylabel("# of Academy Awards",color='green')

# text at the top of every bar the count
for i in range(len(num_oscars)):
    plt.text(i,num_oscars[i] + 0.1,f'{num_oscars[i]}',ha = 'center',fontsize=8,color='green')
plt.xlabel("Movies",color='green')
plt.title("My Favorite Movies",color='red')

plt.show()
```

```python
from collections import Counter
# using bar chart to visulaize the distribution of data 

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

#using lambda function to demonstrate the divide by 10

decile = lambda grade : grade // 10 * 10

freq = Counter(decile(grade) for grade in grades)

plt.bar([x for x in freq.keys()],freq.values(),8,color = 'pink')

for key , values in freq.items():
    plt.text(key,values + 0.2, f'{values}',ha = 'center', color = 'pink' )

plt.axis([-5,105,0,5])

plt.xticks([10 * i for i in range(11)])

plt.xlabel("Decile")
plt.ylabel("# of Students")

plt.title("Distrubtion of Exame 1 Grades")
plt.show()
```

### Line Charts

```python
variance = [1,2,4,5,16,32,64,128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x + y for x , y in zip(variance, bias_squared)]
xs = [i for i , _ in enumerate(variance)]

# we can make multiple calls to plt.plot
# to show multiple series on the same chart

plt.plot(xs, variance, 'g-', label = 'variance') # green solid line
plt.plot(xs, bias_squared, 'r-.', label = 'bias^2') # red dot-dashed line
plt.plot(xs, total_error, 'b:', label= 'total error') #blue dotted line

# because we've assigned labels to each series
# we can get a legend for free
#loc = 9 means 'top center'

plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()

```

### Scatterplots


```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
plt.xlabel('Friends',color = 'black')
plt.ylabel('Minutes',color = 'red')
plt.title("Daily Minutes vs. Number of Friends")

# annotating labelling points
# plt.annotate('Happy', xy = (64,120)) @ annoting

# plt.text(66,160,"sad") # similar w

for label, x_corr , y_corr in zip(labels, friends , minutes):
    plt.annotate(label,
                 xy = (x_corr,y_corr),
                 xytext=(5,-5),
                 textcoords = 'offset points')
    
    

plt.show()
```
